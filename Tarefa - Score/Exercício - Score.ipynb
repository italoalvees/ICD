```python
df = df[["Title","IMDB Rating","IMDB Votes"]]

df.head()
```

```python
df["Score de Popularidade"] = (df['IMDB Votes'] * df['IMDB Rating'])/df['IMDB Votes'].mean()
```

```python
df = df.dropna()
df = df.sort_values(by="Score de Popularidade", ascending = False)
df[0:40]
```
